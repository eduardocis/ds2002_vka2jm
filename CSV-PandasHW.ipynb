```python
# question 1
import pandas as pd

df = pd.read_csv('acc_players-2324F.csv', skiprows = 1) # import
df.head(5)
```

```python
# question 2.1
total_points = df['PTS'].sum()
total_points
```

```python
# question 2.2
most_minutes = df.nlargest(1, 'MP')
most_minutes
```

```python
# question 2.3
top_5_reb = df.nlargest(5, 'TRB')
top_5_reb
```

```python
# question 3.1
newdf = df[df['MP'] > 500]
newdf
```

```python
# question 3.2
most_ast = newdf.nlargest(1, 'AST')
most_ast
```

```python
# question 3.3
three_ast = newdf.nlargest(3, 'AST')
three_ast
```

```python
# question 3.4
three_blocks = newdf.nlargest(3, 'BLK')
three_blocks
```

```python
# question 4.1
school_points = df.groupby('School')['PTS'].sum().reset_index()
school_points
```

```python
# question 4.2 
school_ast = df.groupby('School')['AST'].sum().reset_index()
school_ast
```

```python
# question 4.3
df_pts = school_points.sort_values('PTS', ascending = False)
df_pts.head(3)
```

```python
# extra credit 5.1
import matplotlib.pyplot as plt

top_5_pts = df.nlargest(5, 'PTS')
top_5_pts.plot(kind = 'bar', x = 'Player', y = 'PTS', color = 'red', legend = False)
plt.title('Top 5 Scorers')
plt.xlabel('Player')
plt.ylabel('Points')
plt.show()
```

```python
# extra credit 5.2
import matplotlib.pyplot as plt

plt.scatter(df['FG%'], df['WS'], color = 'red')
plt.title("Scatter Plot Between FG% and WS")
plt.xlabel("FG%")
plt.ylabel("WS")
plt.show()

cc = df['FG%'].corr(df['WS'])
cc
```
