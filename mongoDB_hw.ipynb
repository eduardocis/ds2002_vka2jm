```python
from pymongo import MongoClient
import pprint

connection_string = 'mongodb+srv://user:<pass>@cluster0.72rgr.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0'
client = MongoClient(connection_string)
db = client['sample_mflix']
collection = db['movies']
```

```python
# exercise 1
first_action = collection.find_one({"genres": "Action"})
movies_after_2k = collection.find({"year": {"$gt": 2000}}).limit(5)
above_eight_point_five = collection.find({"imdb.rating": {"$gt": 8.5}}).limit(5)
both_action_adventure = collection.find({"genres": {"$all": ["Action", "Adventure"]}}).limit(5)

pprint.pprint('-' * 100)

pprint.pprint(first_action['title'])
pprint.pprint('-' * 100)

for element in movies_after_2k:
    pprint.pprint(element['title'])
pprint.pprint('-' * 100)

for element in above_eight_point_five:
    pprint.pprint(element['title'])
pprint.pprint('-' * 100)

for element in both_action_adventure:
    pprint.pprint(element['title'])

pprint.pprint('-' * 100)
```

```python
# exercise 2
comedy_ratings = collection.find({"genres": "Comedy"}).sort("imdb.rating", - 1).limit(5)
drama_year = collection.find({"genres": "Drama"}).sort("year", 1).limit(5)

pprint.pprint('-' * 100)

for element in comedy_ratings:
    pprint.pprint(element['title'])
pprint.pprint('-' * 100)

for element in drama_year:
    pprint.pprint(element['title'])
pprint.pprint('-' * 100)
```

```python
# exercise 3
avg_genre_rating = collection.aggregate([
    {"$unwind": "$genres"},
    {"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
])

director_ratings = collection.aggregate([
    {"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
])

num_movies_per_year = collection.aggregate([
    {"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
    {"$sort": {"_id": 1}}
])

pprint.pprint('-' * 100)

for element in avg_genre_rating:
    pprint.pprint(str(element['_id']) + ": " + str(element['avg_rating']))
pprint.pprint('-' * 100)

for element in director_ratings:
    pprint.pprint(str(element['_id']) + ": " + str(element['avg_rating']))
pprint.pprint('-' * 100)

for element in num_movies_per_year:
    pprint.pprint(str(element['_id']) + ": " + str(element['total_movies']))
```

```python
# exercise 4
collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})
collection.update_many({"genres": "Horror", "imdb.rating": {"$exists": False}}, {"$set": {"imdb.rating": 6.0}})
collection.delete_many({"year": {"$lt": 1950}})
```

```python
# exercise 5
collection.create_index([("title", "text")])
love_mov = collection.find({"$text": {"$search": "love"}})
war_mov = collection.find({"$text": {"$search": "war"}}).sort("imdb.rating", -1).limit(5)

pprint.pprint('-' * 100)

for element in love_mov:
    pprint.pprint(element['title'])
pprint.pprint('-' * 100)

for element in war_mov:
    pprint.pprint(element['title'])
```
